```python
#arranging even and odd in an array
l = list(map(int, input("Enter elements: ").split()))

even = []
odd = []

for i in l:
    if i % 2 == 0:
        even.append(i)
    else:
        odd.append(i)

even.sort(reverse=True)
odd.sort()

result = even + odd

print(result)
```

```python
l = list(map(int, input("Enter elements: ").split()))

l.sort()

result = []

for i in range(len(l)-1, -1, -1):
    if l[i] % 2 == 0:
        result.append(l[i])

for i in range(len(l)):
    if l[i] % 2 != 0:
        result.append(l[i])

print(result)
```

```python
#codeforce 427A
l = list(map(int, input("Enter events: ").split()))

police = 0
untreated = 0

for i in l:
    if i == -1:
        if police > 0:
            police -= 1
        else:
            untreated += 1
    else:
        police += i

print("Untreated crimes =", untreated)
```

```python
l = list(map(int, input("Enter events: ").split()))

police = 0
cases = 0
treated = 0
untreated = 0

for i in l:
    if i == -1:
        cases += 1

        if police > 0:
            police -= 1
            treated += 1
        else:
            untreated += 1

    else:
        police += i

print("Available Police =", police)
print("Total Cases =", cases)
print("Treated Cases =", treated)
print("Untreated Cases =", untreated)
```

```python
#duplicates
l = list(map(int, input("Enter elements: ").split()))

new_list = []

for i in l:
    if i not in new_list:
        new_list.append(i)

new_list.sort()

print("List without duplicates:", new_list)
```

```python
#matchstick

matchstick = {
    0: 6, 1: 2, 2: 5, 3: 5, 4: 4,
    5: 5, 6: 6, 7: 3, 8: 7, 9: 6
}

a, b = map(int, input().split())

c = a + b

total = 0

while c > 0:
    digit = c % 10
    total += matchstick[digit]
    c //= 10

print(total)
```
